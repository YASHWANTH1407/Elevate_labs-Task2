```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(f"C:\\Users\\GAJAWADAYASHWANTH\\Downloads\\cleaned_titanic_data.csv")
```

```python
print(df.describe())
print(df.info())
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Age'].dropna(),bins=30,kde=True)
plt.title("Distribution of Age")
plt.show()
```

```python
sns.boxplot(x='Pclass',y='Fare',data=df)
plt.title("Fare distribution by passenger class")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm',fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df[['Age','Fare','Pclass','Survived']].dropna(),hue='Survived')
plt.suptitle("Pairplot",y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a new 'Sex' column from 'Sex_male'
df['Sex'] = df['Sex_male'].apply(lambda x: 'Male' if x == 1 else 'Female')

# Now plot
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival Count by Gender")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```
